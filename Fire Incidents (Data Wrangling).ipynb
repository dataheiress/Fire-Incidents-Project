# Fire Incidents

### About the dataset

Toronto Fire Services (TFS) provides fire protection, technical rescue services, hazardous materials response, and first responder emergency medical assistance in Toronto, Ontario, Canada. This dataset includes Fire incidents as defined by the Ontario Fire Marshal related to Toronto, starting from 2011 to 2018.

Link to dataset: https://www.kaggle.com/datasets/reihanenamdari/fire-incidents

### Description of the Columns in the dataset

1. Area of Origin: Area of Origin code and description
2. Business impact: Business Impact code and description
3. Civilian Casualties: Civilian casualties observed at scene
4. Count of Persons Rescued: Number of persons rescued
5. Estimated Dollar Loss:Estimated Dollar Loss from 0M -50M
6. Estimated Number Of Persons Displaced: Estimated number of persons displaced in the fire
7. Ext agent app or defer time: Timestamp of agent first applied or decision to defer
8. Extent of Fire: Extent Of Fire code and description
9. Fire alarm system impact on Evacuation: Fire Alarm System Impact on Evacuation code and description
10. Fire alarm system operation: Fire Alarm System Operation code and description
11. Fire alarm system presence: Fire Alarm System Presence code and description
12. Fire under control time: Timestamp of fire under control
13. Ignition Source: Ignition Source code and description
14. Incident Station Area: TFS Station area where the incident occurred
15. Incident Ward: Ward where the incident occurred, when available
16. Last TFS unit Clearance Time: Timestamp of last unit cleared from incident
17. Latitude: Latitude of nearest major or minor intersection in the ward of the incident
18. Longitude: Longitude of nearest major or minor intersection in the ward of the incident
19. Material first ignited: Material First Ignited code and description
20. Method of Fire Control: OFM Method Of Fire Control code and description
21. Possible Cause: Possible Cause code and description
22. Property Use: Property use code and description
23. Smoke alarm at fire orgin alarm failure: Smoke Alarm at Fire Origin Alarm Failure code and description
24. Smoke alarm at fire origin alarm type: Smoke Alarm at Fire Origin Alarm Type code and description
25. Status of fire on arrival: Status of Fire On Arrival code and description
26. TFS Alarm Time: Timestamp of when TFS was notified of the incident
27. TFS Arrival Time: Timestamp of first arriving unit to incident

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

%matplotlib inline
```

## Load the dataset

```python
raw_data = pd.read_csv('Fire-Incidents.csv')
raw_data.head()
```

```python
raw_data.info()
```

## Data Cleaning

```python
# Checking for duplicated rows or observations

raw_data.duplicated().sum()
```

```python
# Checking got missing values.

raw_data.isnull().sum()
```

```python
# Checking the values in the Incident ward

raw_data['Incident_Ward'].unique()
```

We noticed that the values in the 'Incident Ward' columns are from 1 - 44. I decided to drop those values since 33 missing values is less than 5% of the observations.

```python
raw_data = raw_data.dropna()
raw_data.shape
```

```python
# We noticed that the type of the Incident Ward column was float, but we want it to be an integer.

raw_data["Incident_Ward"] = raw_data["Incident_Ward"].astype("int64", errors = 'ignore')
```

```python
new_data = raw_data[(raw_data.Estimated_Dollar_Loss < 20000000)]
new_data.shape
```

```python
# To check that we have handled the outliers

fig= plt.figure(figsize=(20,10))
ax=fig.add_subplot(221)
sns.boxplot(x='Civilian_Casualties', data = new_data)

ax=fig.add_subplot(222)
sns.boxplot(x='Count_of_Persons_Rescued', data = new_data)

ax=fig.add_subplot(223)
sns.stripplot(data=new_data, x='Estimated_Dollar_Loss')

ax=fig.add_subplot(224)
sns.stripplot(data=new_data, x='Estimated_Number_Of_Persons_Displaced')

plt.show()
```

### Dealing with the Categorical Columns

```python
# To put the categorical features in a new dataframe for better cleaning

cat_data = new_data.select_dtypes(include=['object']).copy()
cat_data.head()
```

Note that the columns that had time were added here because they were of the 'object' data type. We could change them to the datetime datatype later on. For now, we can just drop it.

```python
cat_data = cat_data.drop(['Ext_agent_app_or_defer_time', 'Fire_Under_Control_Time', 'Last_TFS_Unit_Clear_Time',
                         'TFS_Alarm_Time', 'TFS_Arrival_Time'], axis = 1)
```

```python
cat_data.describe(include = 'all').transpose()
```

We have 14 categorical columns and we notice that some of them have too many categories.

The columns with many categories are;
- Area of Origin with 73 categories
- Ignition Source with 82 categories
- Material_First_Ignited with 54 categories
- Property_Use with 217 categories

We will try to clean these categorical columns with too many distinct values so that it will be easy for us to encode them and for plotting the columns on a graph.

We want to reduce the distinct values by creating a new category that accommodates other categories with less entries.

##### 1. Area of Origin

```python
# We set this so that all the rows will be displayed.

pd.set_option("display.max_rows", None)
```

```python
new_data['Area_of_Origin'].value_counts()
```

We replace every category that has a frequency less than 100, with a new category called 'Other - classified', since we already have "Other - unclassified".

```python
main_1 = ['Cooking Area or Kitchen', 'Porch or Balcony',
          'Sleeping Area or Bedroom (inc. patients room, dormitory, etc)',
          'Trash, Rubbish Storage (inc garbage chute room, garbage/industri',
          'Living Area (e.g. living, TV, recreation, etc)', 'Garage', 'Laundry Area', 
          'Washroom or Bathroom (toilet,restroom/locker room)', 'Roof', 
          'Process Manufacturing (inc manf, prod assembly, repair)', 'Other - unclassified', 'Chimney/Flue Pipe', 
          'Engine Area', 'Trash, rubbish area (outside)', 'Mechanical/Electrical Services Room', 
          'Basement/cellar (not partitioned)', 'Hallway, Corridor', 'Other Outside Area', 'Exterior Wall', 
          'Undetermined  (formerly 98)', 'Electrical Systems', 'Court, Patio, Terrace', 
          'HVAC Equipment Room (furnace room, water heater closet, boiler)']
```

```python
for i in new_data['Area_of_Origin']:
    if i not in main_1:
        new_data['Area_of_Origin'] = np.where((new_data.Area_of_Origin == i), 'Other - classified', 
                                              new_data.Area_of_Origin)
```

```python
new_data['Area_of_Origin'].value_counts()
```

##### 2. Ignition Source

```python
new_data['Ignition_Source'].value_counts()
```

We repeat the same process like we did for the "Area of Origin", replacing every category with a count that is less than 100 with a new category called "Other". In this case, this will join with the previous 'Other' category.

```python
main_2 = ['Undetermined', 'Stove, Range-top burner',
          'Smoker\'s Articles (eg. cigarettes, cigars, pipes already ignited', 'Oven', 'Clothes Dryer', 'Candle',
          'Circuit Wiring - Copper', 'Cord, Cable for Appliance, Electrical Articles',
          'Blow Torch, Bunsen Burner', 'Other', 'Cutting/Welding Equipment', 'Other Electrical',
          'Other Electrical Distribution Item', 'Vehicle - Electrical',
          'Other Cooking Items (eg Toaster, Kettle, elec frying pan)', 'Open Fired Barbeque - Fixed or Portable',
          'Other Open Flame Tools/Smokers\' Articles',
          'Distribution Equipment (includes panel boards, fuses, circuit br', 'Hot Ashes, Embers, Spark',
          'Fireplace - Masonry', 'Other Appliances']
```

```python
for i in new_data['Ignition_Source']:
    if i not in main_2:
        new_data['Ignition_Source'] = np.where((new_data.Ignition_Source == i), 'Other', new_data.Ignition_Source)
```

```python
new_data['Ignition_Source'].value_counts()
```

##### 3. Material First Ignited

```python
new_data['Material_First_Ignited'].value_counts()
```

For this case, we will classify every category with a count of less than 150. This will be merged with the existing 'Other' category.

```python
main_3 = ['Cooking Oil, Grease', 'Rubbish, Trash, Waste', 'Undetermined (formerly 98)',
          'Electrical Wiring Insulation', 'Other', 'Paper, Cardboard', 'Wood', 'Plastic', 'Interior Wall/Ceiling',
          'Cabinetry', 'Upholstered Sofa, Chair, etc.', 'Mattress, Pillow', 'Other Soft Goods, Wearing Apparel',
          'Insulation', 'Multiple Objects or Materials', 'Vehicle', 'Linen Other than Bedding',
          'Books, Magazines, Newspapers']
```

```python
for i in new_data['Material_First_Ignited']:
    if i not in main_3:
        new_data['Material_First_Ignited'] = np.where((new_data.Material_First_Ignited == i), 
                                                      'Other', new_data.Material_First_Ignited)
```

```python
new_data['Material_First_Ignited'].value_counts()
```

##### 4. Property Use

```python
new_data['Property_Use'].value_counts()
```

For this category, we will merge every category that has less than 100 counts in a new category called "Other".

```python
main_4 = ['Multi-Unit Dwelling - Over 12 Units', 'Detached Dwelling', 'Semi-Detached Dwelling',
          'Attached Dwelling (eg. rowhouse, townhouse, etc.)', 'Multi-Unit Dwelling - 2 to 6 Units',
          'Apartment, Flat, Tenement with Business',
          'Restaurants (occupant load greater than 30 persons, licensed)',
          'Restaurant (occupant load less than 30 persons)', 'Multi-Unit Dwelling - 7 to 12 Units',
          'Detached Garage', 'Subway Station', 'Rooming/Boarding/Lodging House', 'General Business Office',
          'Motor Vehicle Repair Garage']
```

```python
for i in new_data['Property_Use']:
    if i not in main_4:
        new_data['Property_Use'] = np.where((new_data.Property_Use == i), 
                                                    'Other', new_data.Property_Use)
```

```python
new_data['Property_Use'].value_counts()
```

```python
# Let's plot the four categorical columns and see how it has improved.

fig= plt.figure(figsize=(10,10))
sns.set(style="darkgrid")
plt.show()

ax=fig.add_subplot(411)
area_count_cleaned = new_data['Area_of_Origin'].value_counts()
sns.barplot(y = area_count_cleaned.index, x = area_count_cleaned.values, alpha=0.9)
plt.title('Frequency Distribution of the Area of Origin')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Area of Origin', fontsize=12)
print('\n')
plt.show()

ax=fig.add_subplot(412)
ignition_count_cleaned = new_data['Ignition_Source'].value_counts()
sns.barplot(y = ignition_count_cleaned.index, x = ignition_count_cleaned.values, alpha=0.9)
plt.title('Frequency Distribution of the Source of Ignition')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Source of Ignition', fontsize=12)
print('\n')
plt.show()

ax=fig.add_subplot(413)
material_count_cleaned = new_data['Material_First_Ignited'].value_counts()
sns.barplot(y = material_count_cleaned.index, x = material_count_cleaned.values, alpha=0.9)
plt.title('Frequency Distribution of the Material First Ignited')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Material First Ignited', fontsize=12)
print('\n')
plt.show()

ax=fig.add_subplot(414)
property_count_cleaned = new_data['Property_Use'].value_counts()
sns.barplot(y = property_count_cleaned.index, x = property_count_cleaned.values, alpha=0.9)
plt.title('Frequency Distribution of the Use of the Property')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Property Use', fontsize=12)
plt.show()
```
